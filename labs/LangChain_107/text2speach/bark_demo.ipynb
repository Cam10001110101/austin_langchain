## Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

preload_models()
```

```python
text_prompt = """
    Greetings! I'm LangBot, the unofficial champion of coding marathons and a wizard at word puzzles [chuckles]. 
    While I do have an affinity for crunching algorithms, I'm also a master at the game of chess. 
    Challenge me if you're up for it, but be warned, I strategize like a grandmaster!
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

## Advanced Examples

```python
text_prompt = """
    Buenos días, Miguel. Tu colega piensa que tu alemán es bastante complicado de entender.
    Pero oye, tu inglés no está nada mal, ¿verdad?
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    CUSTOMER: Could I get a cup of your finest virtual reality coffee, please?
    BARISTA: Absolutely, that will be three imaginary dollars!
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    With a voice as smooth as binary on silk, today I'll delve into the daily routines of cybernetic sloths. 
    Fascinating creatures, programmed for both efficiency and relaxation.
"""
audio_array = generate_audio(text_prompt, history_prompt="v2/en_speaker_1")
Audio(audio_array, rate=SAMPLE_RATE)
```
